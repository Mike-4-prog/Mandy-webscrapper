```python
# Scrapping a html table and converting with pandas dataframe
import pandas as pd
```

```python
Film_Gross = pd.read_html('https://en.wikipedia.org/wiki/List_of_highest-grossing_films')[0]
```

```python
Film_Gross.head(10)
```

```python
Film_Gross.head(10)
```

```python
Film_Gross['Worldwide gross'] = Film_Gross['Worldwide gross'].str.lstrip('T')
Film_Gross['Worldwide gross'] = Film_Gross['Worldwide gross'].str.lstrip('F')
Film_Gross['Worldwide gross'] = Film_Gross['Worldwide gross'].str.lstrip('8')
Film_Gross['Worldwide gross'] = Film_Gross['Worldwide gross'].str.lstrip('$')


```

```python
Film_Gross.head(48)
```

```python
Film_Gross.rename(columns = {'Worldwide gross':'Worldwide_Gross_Earning(USD)'}, inplace = True)
```
